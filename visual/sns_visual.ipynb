```python
# 데이터 불러오기
import json

path = "../data/blog_data/before_month/2022-08-26-충청남도.json"

with open(path, "r") as f:
    data = json.load(f)
print(data)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = "../data/NanumGothic.ttf"
tags = data["item"]
file_name = data["query"]

word_cloud = WordCloud(
    font_path=font_path,
    width=400,
    height=400,
    background_color="white",
    colormap="YlOrRd",
)

word_cloud = word_cloud.generate_from_frequencies(tags)

fig = plt.figure(figsize=(20, 20), facecolor='white')
plt.imshow(word_cloud)
plt.axis("off")

fig.savefig("../data/output/{0}.png".format(file_name))
```
